```python
# Quick check: confirm dataframe exists and date column is datetime
import pandas as pd

# load if df not in memory
try:
    df
except NameError:
    df = pd.read_csv("sales_data.csv", parse_dates=["Date"])

print("Rows:", len(df))
print(df.dtypes)
df.head()
```

```python
# Setup plotting
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# Line chart: total sales by date
daily_sales = df.groupby("Date")["Sales"].sum().sort_index()

plt.figure(figsize=(9,4))
plt.plot(daily_sales.index, daily_sales.values, marker="o", linestyle="-")
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Bar chart: total sales by product
product_sales = df.groupby("Product")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(7,4))
sns.barplot(x=product_sales.index, y=product_sales.values)
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```

```python
# Histogram: distribution of Sales values
plt.figure(figsize=(7,4))
plt.hist(df["Sales"], bins=8, edgecolor="black")
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
